# Matplotlib - Create Charts in browsers

```python
%matplotlib inline  
from matplotlib import pyplot as plt
```

```python
plt.plot([1,2,3, 4], [2,5,7,9])
plt.show()
```

# Showing Labels on Chart

```python
plt.plot([1,2,3, 4], [2,5,7,9])
plt.title("Rumours Data")
plt.ylabel("This is Y")
plt.xlabel("This is X")
plt.show()
```

# Drawing 2 Lines on Same Plot

```python
from matplotlib import style
```

```python
x = [1, 5, 9]
y = [23, 45, 89]
x2 = [3, 7, 10]
y2 = [5, 8, 9]
```

```python
style.use('ggplot')
plt.plot(x,y)
plt.plot(x2,y2)
plt.show()
```

```python
style.use('ggplot')
plt.plot(x,y, label="1st")
plt.plot(x2,y2, label="2nd")
plt.legend()
plt.show()
```

```python
style.use('ggplot')
plt.plot(x,y, label="1st", linewidth=1)
plt.plot(x2,y2, label="2nd", linewidth=1)
plt.legend()
plt.show()
```

```python
style.use('ggplot')
plt.scatter(x,y, label="1st", linewidth=1)
plt.plot(x2,y2, label="2nd", linewidth=1)
plt.legend()
plt.show()
```

```python
import numpy as np
```

```python
plt.style.use('_mpl-gallery')

# make data:
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

# plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
plt.style.use('_mpl-gallery-nogrid')


# make data
x = [1, 2, 3, 4]
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

# plot
fig, ax = plt.subplots()
ax.pie(x, colors=colors, radius=3, center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

# Bar Chart

```python
categories = ['A', 'B', 'C']
values = [10, 20, 15]

plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.show()
```

# Line Chart

```python
import numpy as np

x = np.arange(10)
y = x**2

plt.plot(x, y, marker='o')
plt.title("Line Chart Example")
plt.show()
```
